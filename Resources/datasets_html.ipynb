```python
import pandas as pd
```

```python
import glob
import os
files = []
datasets = []
# glob.glob() return a list of file name with specified pathname
for file in glob.glob(r"./" + "**/*.csv", recursive=True):
  # print the path name of selected files
    files.append(os.path.join(file).replace(".\\",""))
    print(os.path.join(file).replace(".\\",""))
```

```python
for file in files:
    temp_dict = {}
    df = pd.read_csv(file)
    temp_dict['name'] = file
    temp_dict['table'] = df.to_html().replace('table border="1" class="dataframe"','table id="table_id" class="display')
    datasets.append(temp_dict)
```
